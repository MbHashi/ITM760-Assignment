```python
import pandas as pd
df = pd.read_csv("https://github.com/tofighi/ITM760/blob/main/dataset/news_popularity.csv?raw=true")
df.head(10)
```

## Task 1

Report the size of dataset (number of columns and rows), clean it (drop non-predictive columns), and create scatter matrix (similar to labs) for an optional subset (for example choose 10 numerical attributes) of numerical attributes.

```python
#Reporting the size of the dataset:
print(df.shape, "\n")

#Describing the dataset:
df.info()

#Dropping non-predictive columns:
df.drop(labels=['url', 'timedelta'], axis=1, inplace=True)
```

```python
#Creating a Scatter Matrix for Correlation Analysis:

#Importing Scatter Matrix from the Pandas library:
import pandas as pd 
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
from matplotlib import cm


#Best 10 numerical attributes:
numeric = ['n_tokens_title', 'n_tokens_content', 'n_unique_tokens', 
           'num_hrefs', 'num_self_hrefs', 'num_imgs', 
           'num_videos', 'num_keywords', 'shares']                       
X = df[numeric]
y = df['shares']
cmap = cm.get_cmap('gnuplot')
scatter = scatter_matrix(X, c = y, marker = 'o', s=40, 
                         hist_kwds={'bins':15}, figsize=(9,9), cmap = cmap)
plt.suptitle('Scatter-matrix for each input variable')
plt.savefig('Numeric_Scatter_Matrix')
```

```python
#Exploring the data:
import seaborn as sns

sns.kdeplot(df['shares'], color="red")
#The data is right skewed with a lower bound
```

## Task 2

Split the dataset into a training set (70%) and a test set (30%) and apply scaling, etc (similar to lab)

```python
#Create Training and Test Sets splits:
#Applying Scaling:
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
print("Data sets are split and scaled!")
```

## Task 3

```python
#Training a KNN model where k=3:
#Overall Run-time for all of the KNN functions:
from datetime import datetime
start_time = datetime.now()

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
print('Accuracy of K-NN=3 classifier on training set: {:.2f}'
     .format(knn.score(X_train, y_train)))
print('Accuracy of K-NN=3 classifier on test set: {:.2f}'
     .format(knn.score(X_test, y_test)))


#===================================================================
print("\n========================================================\n")

#Training a KNN model where k=5:

from sklearn.neighbors import NearestNeighbors
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
print('Accuracy of K-NN=5 classifier on training set: {:.2f}'
     .format(knn.score(X_train, y_train)))
print('Accuracy of K-NN=5 classifier on test set: {:.2f}'
     .format(knn.score(X_test, y_test)))


#===================================================================
print("\n========================================================\n")

#Training a KNN model where k=7:

from sklearn.neighbors import NearestNeighbors
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train, y_train)
print('Accuracy of K-NN=7 classifier on training set: {:.2f}'
     .format(knn.score(X_train, y_train)))
print('Accuracy of K-NN=7 classifier on test set: {:.2f}'
     .format(knn.score(X_test, y_test)))

#===================================================================
print("\n========================================================\n")

#Training a KNN model where k=9:

from sklearn.neighbors import NearestNeighbors
knn = KNeighborsClassifier(n_neighbors=9)
knn.fit(X_train, y_train)
print('Accuracy of K-NN=9 classifier on training set: {:.2f}'
     .format(knn.score(X_train, y_train)))
print('Accuracy of K-NN=9 classifier on test set: {:.2f}'
     .format(knn.score(X_test, y_test)))

#===================================================================
print("\n========================================================\n")

#Training a KNN model where k=11:

from sklearn.neighbors import NearestNeighbors
knn = KNeighborsClassifier(n_neighbors=11)
knn.fit(X_train, y_train)
print('Accuracy of K-NN=11 classifier on training set: {:.2f}'
     .format(knn.score(X_train, y_train)))
print('Accuracy of K-NN=11 classifier on test set: {:.2f}'
     .format(knn.score(X_test, y_test)))

#===================================================================
print("\n========================================================\n")

end_time = datetime.now()
print('The overall duration for the KNN is: {}'.format(end_time - start_time))
```

## Task 4
Train SVM similar to the lab document and calculate runtime and accuracy

```python
#Support Vector Machine:
#The Run-time for the SVM model:
start_time = datetime.now()

from sklearn.svm import SVC
svm = SVC()
svm.fit(X_train, y_train)
print('Accuracy of SVM classifier on training set: {:.2f}'
     .format(svm.score(X_train, y_train)))
print('Accuracy of SVM classifier on test set: {:.2f}'
     .format(svm.score(X_test, y_test)))

#===================================================================
print("\n========================================================\n")

end_time = datetime.now()
print('The overall duration for the SVM model is: {}'
      .format(end_time - start_time))
```

```python
#The Confusion matrix for SVM model:

from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
pred = svm.predict(X_test)
print(confusion_matrix(y_test, pred))
print(classification_report(y_test, pred))
```

## Task 5

Train Decision Tree similar to the lab document and calculate runtime and accuracy
